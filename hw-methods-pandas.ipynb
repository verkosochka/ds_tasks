3.1. Створіть нову колонку в даних `Rating Category`, яка визначає категорію рейтингу: високий, середній чи низький за наступним правилом:
- якщо рейтинг вищий або рівний 8 - то він високий (`High`)
- якщо рейтинг від 6 до 8 не включаючи - то він середній (`Medium`)
- інакше - низький (`Low`).

```python
def category_mapping(value: Union[int, float]) -> str:
    """
    Map a numerical value to a categorical label based on predefined ranges.

    Parameters
    ----------
    value : Union[int, float]
        The numerical value to categorize.

    Returns
    -------
    str
        The categorical label for the given value. Possible values are:
        - 'High' for values >= 8
        - 'Medium' for values >= 6 and < 8
        - 'Low' for values < 6

    Examples
    --------
    >>> category_mapping(9)
    'High'

    >>> category_mapping(7)
    'Medium'

    >>> category_mapping(5)
    'Low'
    """
    if value >= 8:
        return 'High'
    elif value >= 6:
        return 'Medium'
    else:
        return 'Low'
```

```python
supermarket_dataset['Rating Category'] = supermarket_dataset.Rating.apply(category_mapping)
supermarket_dataset.head()
```

```python
supermarket_dataset['Rating Category'] = supermarket_dataset['Rating Category'].astype('category')
```

4.1. Запустіть код нижче аби створити змінну `time`. Ця змінна містить час в форматі "година:хвилина". Напишіть функцію `get_hour`, яка буде зі змінної формату як `time` дістати лише години і конвертувати їх у тип `int`.

Запустіть написану функцію на змінній `time` та перевірте правильність роботи функції.

```python
time = '11:28'
```

```python
def get_hour(time: str) -> int:
    """
    Extract the hour from a time string in the format "HH:MM:SS" or "HH:MM".

    Parameters
    ----------
    time : str
        The time string in the format "HH:MM:SS" or "HH:MM".

    Returns
    -------
    int
        The hour extracted from the time string as an integer.

    Examples
    --------
    >>> get_hour("14:30:00")
    14

    >>> get_hour("08:45")
    8
    """
    return int(time.split(':')[0])

get_hour(time)
```

4.2. З допомогою функції `get_hour` і метода `DataFrame.apply()` створіть нову колонку `Hour` в нашому наборі даних, яка буде містити години здійснення покупки, обчислені на основі часу здійснення покупки `Time`.

```python
supermarket_dataset['Hour'] = supermarket_dataset.Time.apply(get_hour)
supermarket_dataset.head()
```
