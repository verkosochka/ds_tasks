1. Імпортуйте необхідні бібліотеки:


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```


2. Використовуючи фукнціонал `numpy` створіть масив `x` з 400 точок в діапазоні між -2 та 2 і масив `x_positive` в діапазоні між 0.1 і 2 також з 400 точок. Другий масив в нас буде для логарифмічної функції, перший з більш широким діапазоном - для інших функцій.


```python
x = np.linspace(-2, 2, 400)  # array implementation for exponential and modified sin functions
x_positive = np.linspace(0.1, 2, 400) # array for logarifmic function
```


3. Тут маємо кілька кроків.

  3.1. Ініціалізуйте фігуру Matplotlib з розміром, достатнім для трьох панелей (рекомендую спробувати експериментально виявити добрий розмір).   
  3.2. На першій панелі візуалізуйте експоненціальну функцію `y = e^x`. Використовуйте відповідну змінну `x`.  
  3.2. На другій панелі покажіть логарифмічну функцію `y = ln(x)`.  
  3.4. На третій панелі покажіть модифіковану синусоїду `y = sin(x) + x`.  

  Відобразіть графік.


```python
fig, ax = plt.subplots(3, figsize=(8, 15))
plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'serif' 
plt.axis('equal')

ax[0].plot(x, np.exp(x))
ax[0].grid(True, linestyle='--', linewidth=0.5)
ax[0].set_title(r"Exponential Function $f(x) = e^x$")

ax[1].plot(x_positive, np.log(x_positive))
ax[1].grid(True, linestyle='--', linewidth=0.5)
ax[1].set_title(r"Logarithmic Function $f(x) = \ln(x)$")

ax[2].plot(x, np.sin(x) + x)
ax[2].grid(True, linestyle='--', linewidth=0.5)
ax[2].set_title(r"Sine Function with Linear Growth $f(x) = \sin(x) + x$");
```

4. Модифікація і довершення графіку.  
  4.1. Для кожної панелі встановіть назву графіка та підписи до осей х та у (можна так і назвати `x` та формула функції).  
  4.2. Використайте різні кольори для кожного графіка.  


```python
fig, ax = plt.subplots(3, figsize=(8, 15))

# LaTeX and font settings
plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'serif'

# region --- Plot 1: Exponential Function ---
ax[0].plot(x, np.exp(x), 
           color='green')

ax[0].grid(True, 
           linestyle='--', 
           linewidth=0.3)  

ax[0].set_xlabel('x', 
                 fontsize=14) 

ax[0].set_ylabel(r'$f(x) = e^x$', 
                 fontsize=14) 

ax[0].set_title(r"Exponential Function $f(x) = e^x$", 
                fontsize=16)
# endregion

# region --- Plot 2: Logarithmic (Natural Logarithm) Function ---
ax[1].plot(x_positive, np.log(x_positive), 
           color='red')

ax[1].grid(True, 
           linestyle='--', 
           linewidth=0.3) 

ax[1].set_xlabel('x', 
                 fontsize=14) 

ax[1].set_ylabel(r'$f(x) = \ln(x)$', 
                 fontsize=14) 

ax[1].set_title(r"Logarithmic Function $f(x) = \ln(x)$", 
                fontsize=16)  

# endregion 

# region --- Plot 3: Sine Function with linear growth ---
ax[2].plot(x, np.sin(x) + x, 
           color='darkblue')

ax[2].grid(True, linestyle='--', 
           linewidth=0.3) 

ax[2].set_xlabel('x', 
                 fontsize=14)  

ax[2].set_ylabel(r'$f(x) = \sin(x) + x$',
                 fontsize=14) 

ax[2].set_title(r"Sine Function with Linear Growth $f(x) = \sin(x) + x$", 
                fontsize=16)  
# endregion 

# Default settings 
fig.tight_layout()
plt.axis('equal')
plt.show()
```

```python
# If we want to optimize our code, we can use OOP concepts to automize each visualization using list of data dictionaries and loops

fig, ax = plt.subplots(3, figsize=(8, 15))

plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'serif'

plots = [
    {'data': (x, np.exp(x)), 'color': 'green', 
     'xlabel': 'x', 'ylabel': r'$f(x) = e^x$', 
     "title": r"Exponential Function $f(x) = e^x$"} ,
    
    {'data': (x_positive, np.log(x_positive)), 'color': 'red', 
     'xlabel': 'x', 'ylabel': r'$f(x) = \ln(x)$', 
     'title': r'Logarithmic Function $f(x) = \ln(x)$'},
    
    {'data': (x, np.sin(x) + x), 'color': 'darkblue', 
     'xlabel': 'x', 'ylabel': r'$f(x) = \sin(x) + x$', 
     'title': r'Sine Function with Linear Growth $f(x) = \sin(x) + x$'}
]

for i, plot in enumerate(plots):
    ax[i].plot(*plot['data'], color=plot['color'])
    ax[i].grid(True, linestyle='--', linewidth=0.3)
    ax[i].set_xlabel(plot['xlabel'], fontsize=14)
    ax[i].set_ylabel(plot['ylabel'], fontsize=14)
    ax[i].set_title(plot['title'], fontsize=16)

fig.tight_layout()
plt.show()
```

5. А тепер зобразіть ті самі 3 трафіки тих самих кольорів тільки на одному графіку. Визначіть назву кожної з ліній для легенди та обовʼязково відобразіть легенду.

```python
plt.plot(x, np.exp(x), 
         '-g', 
         label=r'$f(x) = e^x$')
plt.plot(x_positive, np.log(x_positive), 
         '-r', 
         label=r'$f(x) = \ln(x)$')
plt.plot(x, np.sin(x) + x, 
         '-b', 
         label=r'$f(x) = \sin(x) + x$')

plt.axhline(0, 
            color='black', 
            linewidth=0.5, 
            linestyle='--')  
plt.axvline(0, 
            color='black', 
            linewidth=0.5, 
            linestyle='--')  

equation_1 = 'Exponential function'
equation_2 = 'Logarithmic function'
equation_3 = 'Sine function with linear growth'

plt.text(2, 4, equation_1)
plt.text(0.75, -1, equation_2)
plt.text(1, 1.25, equation_3)

plt.axis('equal')
plt.title('Basic functions and transformations')
plt.xlabel('$x$', fontsize=12)
plt.ylabel('$f(x)$', fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.3, alpha=0.7)
plt.axis('equal')
plt.legend()
plt.savefig('my_figure.png'); # Task 6
```
