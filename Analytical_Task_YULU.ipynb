```python
# importing libraries and packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# Set default LaTex and font settings
plt.rcParams['font.family'] = 'serif'
plt.rcParams['text.usetex'] = True
```

```python
# Reading dataset file
path = 'data/yulu_rental.csv'
yulu_rental_df = pd.read_csv(path)

yulu_rental_df.head()
```

```python
total_bytes = yulu_rental_df.memory_usage().sum()
total_megabytes = total_bytes / (1024 * 1024)
formatted_output = f'{total_megabytes:.2f} MB'
print(f'For non-technical users this dataset occupies {formatted_output} of RAM. The raw output of calculation is {total_megabytes}.')
```

```python
yulu_rental_df.info()
```

```python
# Fature'daytype'
# 2: holiday, 1: weekend, 0: working day
conditions = [
    yulu_rental_df['holiday'] == 1,  # holiday
    yulu_rental_df['workingday'] == 0,  # weekend
    yulu_rental_df['workingday'] == 1   # working day
]

choices = [2, 1, 0]  # 2: holiday, 1: weekend, 0: wotking day

# New column 'daytype'
yulu_rental_df['daytype'] = np.select(conditions, choices)
yulu_rental_df.head(18)
```

```python
total_bytes = yulu_rental_df.memory_usage().sum()
total_megabytes = total_bytes / (1024 * 1024)
formatted_output = f'{total_megabytes:.2f} MB'
print(f'For non-technical users this dataset occupies {formatted_output} of RAM. The raw output of calculation is {total_megabytes}.')
```

<h1 style='font-family: Garamond;'>Boxplot діаграми, які порівнюють кількість оренд велосипедів серед <b>casual</b> (незареєстрованих) та <b>registered</b> (зареєстрованих) користувачів у різні сезони (Spring, Summer, Autumn, Winter). Розділення за типами днів (Day Type)</h1>

```python
yulu_rental_df['season_name'] = yulu_rental_df['season'].map({
    1: 'Spring', 2: 'Summer', 3: 'Autumn', 4: 'Winter'
})

fig, axes = plt.subplots(1, 2, figsize=(14, 7), sharey=True)

sns.boxplot(data=yulu_rental_df, x='season_name', y='casual', hue='daytype',
            palette='pastel', ax=axes[0])
axes[0].set_title('Casual Users Across Seasons', fontsize=14)
axes[0].set_xlabel('Season', fontsize=12)
axes[0].set_ylabel('Number of Rentals', fontsize=12)
axes[0].grid(True, linestyle='--', alpha=0.5)
axes[0].legend(title='Day Type', loc='upper right')

sns.boxplot(data=yulu_rental_df, x='season_name', y='registered', hue='daytype',
            palette='pastel', ax=axes[1])
axes[1].set_title('Registered Users Across Seasons', fontsize=14)
axes[1].set_xlabel('Season', fontsize=12)
axes[1].set_ylabel('')  
axes[1].grid(True, linestyle='--', alpha=0.5)
axes[1].legend(title='Day Type', loc='upper right')

fig.suptitle('Bike Rentals Across Seasons for Casual and Registered Users', fontsize=16)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show();
```

```python
yulu_rental_df['hour'] = pd.to_datetime(yulu_rental_df['datetime']).dt.hour

hourly_rentals = yulu_rental_df.groupby('hour')['count'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(hourly_rentals['hour'], hourly_rentals['count'], marker='o', linestyle='-', color='blue')
plt.title('Average Bike Rentals by Hour of the Day', fontsize=14)
plt.xlabel('Hour of the Day', fontsize=12)
plt.ylabel('Average Number of Rentals', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(range(0, 24))  
plt.show()
```

<h1 style='font-family: Garamond;'>Cередня кількість оренд велосипедів за годинами протягом дня для <b>casual</b> користувачів (помаранчева лінія) та <b>registered</b> користувачів (синя лінія)</h1>

```python
# Створення колонки "hour" з часу доби
yulu_rental_df['hour'] = pd.to_datetime(yulu_rental_df['datetime']).dt.hour

# Агрегація даних за годинами
hourly_data = yulu_rental_df.groupby('hour')[['casual', 'registered']].mean()

# Побудова графіку
plt.figure(figsize=(10, 6))
plt.plot(hourly_data.index, hourly_data['casual'], label='Casual Users', color='orange', marker='o')
plt.plot(hourly_data.index, hourly_data['registered'], label='Registered Users', color='blue', marker='o')

# Налаштування графіку
plt.title('Average Bike Rentals by Hour for Casual and Registered Users', fontsize=14)
plt.xlabel('Hour of the Day', fontsize=12)
plt.ylabel('Average Number of Rentals', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show();
```

```python
peak_hours = hourly_rentals.sort_values(by='count', ascending=False)
print("Peak hours with highest average bike rentals:")
peak_hours.head(7)
```

<h1 style='font-family: Garamond;'><b>Аналіз 3D графіку</b>: уплив швидкості вітру та вологості на кількість оренд у пікові години</h1>

```python
from mpl_toolkits.mplot3d import Axes3D

yulu_rental_df['hour'] = pd.to_datetime(yulu_rental_df['datetime']).dt.hour
peak_hours_df = yulu_rental_df[(yulu_rental_df['hour'] >= 7) & (yulu_rental_df['hour'] <= 9) | 
                               (yulu_rental_df['hour'] >= 17) & (yulu_rental_df['hour'] <= 19)]

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(peak_hours_df['windspeed'], peak_hours_df['humidity'], peak_hours_df['count'], 
           c=peak_hours_df['count'], cmap='coolwarm', alpha=0.7)

ax.set_title('Impact of Windspeed and Humidity on Bike Rentals in Peak Hours', fontsize=14)
ax.set_xlabel('Windspeed', fontsize=10, labelpad=1)
ax.set_ylabel('Humidity', fontsize=10, labelpad=1)
ax.set_zlabel('Number of Rentals', fontsize=10, labelpad=1)

plt.show();
```

<h1 style='font-family: Garamond;'>Pairplot взаємозв'язків між кількома числовими ознаками: <b>atemp</b> (відчутна температура), <b>humidity</b> (вологість), <b>windspeed</b> (швидкість вітру) та <b>rentals</b> (кількість оренд). Дані розділено за <b>User Type</b> — casual (незареєстровані користувачі, червоні) та registered (зареєстровані користувачі, сині)</h1>

```python
selected_columns = ['temp', 'atemp', 'humidity', 'windspeed', 'casual', 'registered']
pairplot_df = yulu_rental_df[selected_columns]

pairplot_df_melted = pairplot_df.melt(
    id_vars=['temp', 'atemp', 'humidity', 'windspeed'],
    value_vars=['casual', 'registered'],
    var_name='User_Type',
    value_name='Rentals'
)

sns.pairplot(data=pairplot_df_melted, hue='User_Type', palette='Set1',
             vars=['atemp', 'humidity', 'windspeed', 'Rentals'], diag_kind='kde',  plot_kws={'alpha': 0.5})

plt.suptitle('Pairplot Analysis: Influence of Weather on Casual vs Registered Rentals', 
             y=1.02, fontsize=16)
plt.show();
```

<h1 style='font-family: Garamond;'>Аналіз впливу вологості на оренди <b>casual</b> та <b>registered</b> користувачів у пікові години</h1>

```python
yulu_rental_df['hour'] = pd.to_datetime(yulu_rental_df['datetime']).dt.hour

peak_hours_df = yulu_rental_df[(yulu_rental_df['hour'] >= 7) & (yulu_rental_df['hour'] <= 9) | 
                               (yulu_rental_df['hour'] >= 17) & (yulu_rental_df['hour'] <= 19)]

fig, axes = plt.subplots(1, 2, figsize=(16, 8))

sns.scatterplot(data=peak_hours_df, x='humidity', y='casual', hue='hour', palette='viridis', ax=axes[0], alpha=0.7)
axes[0].set_title('Impact of Humidity on Casual Rentals (Peak Hours)', fontsize=14)
axes[0].set_xlabel('Humidity')
axes[0].set_ylabel('Casual Rentals')
axes[0].grid(True, linestyle='--', alpha=0.5)

sns.scatterplot(data=peak_hours_df, x='humidity', y='registered', hue='hour', palette='viridis', ax=axes[1], alpha=0.7)
axes[1].set_title('Impact of Humidity on Registered Rentals (Peak Hours)', fontsize=14)
axes[1].set_xlabel('Humidity')
axes[1].set_ylabel('Registered Rentals')
axes[1].grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show();
```
