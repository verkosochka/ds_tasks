````python
"""
Module for circle calculations.

This module provides functions to calculate the area and circumference of a circle.

Functions
---------
calc_area(r: float) -> float
    Calculate the area of a circle with radius r.
calc_circumference(r: float) -> float
    Calculate the circumference of a circle with radius r.
"""

from math import pi
from typing import Union

# Define a type alias for allowed numeric inputs
Number = Union[int, float]

def calc_area(r: Number) -> float:
    """
    Calculate the area of a circle.

    Parameters
    ----------
    r : int | float
        The radius of the circle.

    Returns
    -------
    float
        The area of the circle calculated as pi * r ** 2.

    Raises
    ------
    ValueError
        If the radius is negative.
    """
    if r < 0:
        raise ValueError("Radius cannot be negative.")
    
    return pi * r ** 2


def calc_circumference(r: Number) -> float:
    """
    Calculate the circumference of a circle.

    Parameters
    ----------
    r : int | float
        The radius of the circle.

    Returns
    -------
    float
        The circumference of the circle calculated as 2 * pi * r.

    Raises
    ------
    ValueError
        If the radius is negative.
    """
    if r < 0:
        raise ValueError("Radius cannot be negative.")
    
    return 2 * pi * r


if __name__ == "__main__":
    radius: int = 5  # Explicit type hint for clarity
    print(f"Area: {calc_area(radius)}, Circumference: {calc_circumference(radius)}")
````

````python
import numpy as np
import pandas as pd

def main():
    """
    Create a large DataFrame representing store transactions, convert specified columns to
    categorical type for memory efficiency, and measure memory usage before and after conversion.

    The DataFrame includes 1 million rows with the following columns:
        - product_name: Name of the product.
        - product_category: Category of the product.
        - product_price: Price of the product.
        - transaction_date: Date and time of the transaction.

    After conversion, the code prints:
        - Memory usage before conversion (in KB)
        - Memory usage after conversion (in KB)
        - The factor by which the memory usage is reduced.
    
    Returns
    -------
    None
    """
    np.random.seed(0)

    products = ['apple', 'banana', 'cherry', 'dates', 'elderberry']
    categories = ['fruit', 'fruit', 'fruit', 'fruit', 'fruit']

    N_ROWS = 10**6
    df = pd.DataFrame({
        'product_name': np.random.choice(products, N_ROWS),
        'product_category': np.random.choice(categories, N_ROWS),
        'product_price': np.random.uniform(1, 10, N_ROWS),
        'transaction_date': pd.date_range(start='1/1/2000', periods=N_ROWS, freq='h')
    })

    # Measure memory usage before conversion (in bytes)
    memory_before = df.memory_usage(deep=True).sum()

    # Convert 'product_name' and 'product_category' columns to categorical type
    df['product_name'] = df['product_name'].astype('category')
    df['product_category'] = df['product_category'].astype('category')

    # Measure memory usage after conversion (in bytes)
    memory_after = df.memory_usage(deep=True).sum()

    print(f"Memory usage before conversion: {memory_before / 1024:.2f} KB")
    print(f"Memory usage after conversion: {memory_after / 1024:.2f} KB")
    print(f"Memory reduction factor: {memory_before / memory_after:.2f}x")


if __name__ == '__main__':
    main()
````

````python
from typing import List, Union

def calculate_variance(numbers: List[Union[int, float]]) -> float:
    """
    Calculate the variance of a list of numbers.

    The variance is computed using the formula:
        variance = mean(x^2) - (mean(x))^2
    where mean(x) is the average of the numbers and mean(x^2) is the average of the squares.

    Args:
        numbers (List[Union[int, float]]): A list of numerical values.

    Returns:
        float: The variance of the numbers.

    Raises:
        ValueError: If the input list is empty.
    """
    if not numbers:
        raise ValueError("Input list cannot be empty.")

    n = len(numbers)
    mean = sum(numbers) / n
    mean_squares = sum(x * x for x in numbers) / n
    variance = mean_squares - mean ** 2
    return variance
````

````python
import numpy as np

def calculate_variance_np(numbers: List[Union[int, float]]) -> float:
    if not numbers:
        raise ValueError("Input list cannot be empty.")
    arr = np.array(numbers)
    mean = np.mean(arr)
    mean_squares = np.mean(arr ** 2)
    return mean_squares - mean ** 2
````
